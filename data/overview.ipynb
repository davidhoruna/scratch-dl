```python
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
import numpy as np
import os

path = "fruits/fruits-360_100x100/fruits-360/Training"
fruits = os.listdir(path)
len(fruits)
```

```python

selected = np.random.choice(fruits,4)
nrows = 4
ncols = 2
fig, ax =plt.subplots(nrows, ncols, figsize=(10,10))

i = 0
r=0
c=0

while r < nrows and c < ncols:

    for fruit in selected:
        fruits_dir = os.path.join(path, fruit)
        fruits_paths = os.listdir(fruits_dir)
        img_paths = np.random.choice(fruits_paths, 2, replace=False)
        for img_path in img_paths:
            img=plt.imread(os.path.join(fruits_dir, img_path))
            ax[r,c].imshow(img)
            ax[r,c].axis('off')
            ax[r,c].set_title(fruit)
            c+=1
        r+=1
        c=0

            
plt.show()
    
    
```
